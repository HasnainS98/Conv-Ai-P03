Imports necessary libraries:

	•	WordCloud and STOPWORDS from wordcloud for creating word clouds and managing common stopwords.
	•	matplotlib.pyplot as plt for plotting.
	•	numpy as np for handling numerical operations.
	•	Image from PIL (Python Imaging Library) for loading image files used as a mask.

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

	•	Opens and reads the contents of the JaneEyre.txt file located in Google Drive.
	•	Stores the entire text in the text variable, which will be used to generate the word cloud.

```python
with open("/content/drive/My Drive/Colab Notebooks/JaneEyre.txt", "r") as file:
   text= file.read()
```

	•	Specifies the path to a custom mask image located in Google Drive.
	•	Loads this image and converts it to a NumPy array (mask_image), enabling it to be used as a shape for the word cloud.

```python
mask_path = "//content/drive/My Drive/Colab Notebooks/360_F_240635575_EJifwRAbKsVTDnA3QE0bCsWG5TLhUNEZ.jpg"
mask_image = np.array(Image.open(mask_path))
```

Creates a customized word cloud using the WordCloud class with specified parameters:

	•	width and height: Dimensions of the word cloud.
	•	background_color: Sets a light blue background.
	•	colormap: Defines a “plasma” color map for the word cloud.
	•	max_words: Limits the word cloud to the 150 most frequent words.
	•	stopwords: Uses default stopwords to exclude common words.
	•	mask: Applies the custom mask shape from the mask_image.
	•	contour_color and contour_width: Adds a black contour with a specified width around the shape.
	•	Generates the word cloud based on the text variable.

```python
wordcloud_custom = WordCloud(
    width=800,
    height=400,
    background_color="lightblue",  # Custom background color
    colormap="plasma",  # Custom color map
    max_words=150,  # Limit to 150 words
    stopwords=set(STOPWORDS),  # Using default stopwords
    mask=mask_image,  # Using the mask for custom shape
    contour_color="black",
    contour_width=1
).generate(text)
```

	Plots the generated word cloud:
  
	•	Sets the figure size to 10x10 inches.
	•	Displays the word cloud image without grid lines (axis("off")).
	•	Renders the plot using plt.show().

```python
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud_custom, interpolation="bilinear")
plt.axis("off")
plt.show()
```
